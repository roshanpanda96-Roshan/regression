```python
# Linear Regression using Mean Method (Least Squares)

# Input X values
x_input = input("Enter X values separated by space: ").split()
x = []
for val in x_input:
    x.append(float(val))

# Input Y values
y_input = input("Enter Y values separated by space: ").split()
y = []
for val in y_input:
    y.append(float(val))

# Check length
if len(x) != len(y):
    print("Error: X and Y must have the same number of values.")


def linear_regression_mean_method(x, y):
    n = len(x)

    # Compute means
    x_mean = sum(x) / n
    y_mean = sum(y) / n

    # Compute slope (b)
    numerator = sum((x[i] - x_mean) * (y[i] - y_mean) for i in range(n))
    denominator = sum((x[i] - x_mean) ** 2 for i in range(n))
    b = numerator / denominator

    # Compute intercept (a)
    a = y_mean - b * x_mean

    return a, b

a, b = linear_regression_mean_method(x, y)

print("Intercept (a):", a)
print("Slope (b):", b)

# Predict example
def predict(x_value):
    return a + b * x_value

print("Prediction for x=6:", predict(6))
```
